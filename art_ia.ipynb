```python
import face_recognition as fr
import os
import cv2
import face_recognition
import numpy as np
# from time import sleep

def get_encoded_faces():
    encoded = {}

    for dirpath, dnames, fnames in os.walk("./faces"):
        for f in fnames:
            if f.endswith(".jpg") or f.endswith(".png"):
                face = fr.load_image_file("faces/" + f)
                encoding = fr.face_encodings(face)[0]
                encoded[f.split(".")[0]] = encoding

    return encoded

def unknown_image_encoded(img):
    face = fr.load_image_file("faces/" + img)
    encoding = fr.face_encodings(face)[0]

    return encoding
```

```python
import cv2

# Load the pre-trained face detection model
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Load the image
img = cv2.imread('/content/drive/MyDrive/WillSolo.jpg')

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect faces
faces = face_cascade.detectMultiScale(gray, 1.1, 4)

# Draw rectangles around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)

# Display the output
cv2.imshow('img', img)
cv2.waitKey()
```

```python
import cv2
import face_recognition
from google.colab.patches import cv2_imshow  # Para mostrar imágenes en Google Colab

# Imagen de referencia (persona a reconocer)
image = cv2.imread("/content/drive/MyDrive/WillSolo.jpg")

# Detectar la ubicación de la cara en la imagen de referencia
face_loc = face_recognition.face_locations(image)[0]
face_image_encodings = face_recognition.face_encodings(image, known_face_locations=[face_loc])[0]

# Mostrar la imagen con un rectángulo alrededor de la cara de la persona
cv2.rectangle(image, (face_loc[3], face_loc[0]), (face_loc[1], face_loc[2]), (0, 255, 0))
cv2_imshow(image)  # Mostrar imagen de referencia

# Cargar la imagen grupal
group_image = cv2.imread("/content/drive/MyDrive/WillGrupo.jpg")

# Detectar todas las caras en la imagen grupal
group_face_locs = face_recognition.face_locations(group_image)
group_face_encodings = face_recognition.face_encodings(group_image, known_face_locations=group_face_locs)

# Comparar las codificaciones de la cara de referencia con las caras de la imagen grupal
for (top, right, bottom, left), face_encoding in zip(group_face_locs, group_face_encodings):
    matches = face_recognition.compare_faces([face_image_encodings], face_encoding)

    # Si se reconoce la cara, dibujar un rectángulo alrededor de la cara en la imagen grupal
    if matches[0]:
        cv2.rectangle(group_image, (left, top), (right, bottom), (0, 255, 0), 2)
    else:
        cv2.rectangle(group_image, (left, top), (right, bottom), (255, 0, 0), 2)

# Mostrar la imagen grupal con las caras reconocidas
cv2_imshow(group_image)
```

```python
import cv2
import face_recognition
from google.colab.patches import cv2_imshow  # Para mostrar imágenes en Google Colab

# Imagen de referencia (persona a reconocer)
image = cv2.imread("/content/drive/MyDrive/WillSolo.jpg")

# Detectar la ubicación de la cara en la imagen de referencia
face_loc = face_recognition.face_locations(image)[0]
face_image_encodings = face_recognition.face_encodings(image, known_face_locations=[face_loc])[0]

# Mostrar la imagen de referencia con un rectángulo alrededor de la cara
cv2.rectangle(image, (face_loc[3], face_loc[0]), (face_loc[1], face_loc[2]), (0, 265, 0), 2)
cv2_imshow(image)  # Mostrar imagen de referencia

# Cargar la imagen grupal
group_image = cv2.imread("/content/drive/MyDrive/WillGrupo.jpg")

# Detectar todas las caras en la imagen grupal
group_face_locs = face_recognition.face_locations(group_image)
group_face_encodings = face_recognition.face_encodings(group_image, known_face_locations=group_face_locs)

# Nombre de la persona en la primera imagen
name_of_known_person = "Will"

# Comparar las codificaciones de la cara de referencia con las caras de la imagen grupal
for (top, right, bottom, left), face_encoding in zip(group_face_locs, group_face_encodings):
    # Compara la cara con la codificación de la persona conocida
    matches = face_recognition.compare_faces([face_image_encodings], face_encoding)

    # Si se reconoce la cara, poner el nombre; de lo contrario, poner "Desconocido"
    if matches[0]:
        label = name_of_known_person
        color = (0, 255, 0)  # Verde para la persona reconocida
    else:
        label = "Desconocido"
        color = (255, 0, 0)  # azul para desconocidos

    # Dibujar un rectángulo alrededor de la cara
    cv2.rectangle(group_image, (left, top), (right, bottom), color, 2)

    # Poner la etiqueta encima del rectángulo
    cv2.putText(group_image, label, (left, top - 8), cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2)

# Mostrar la imagen grupal con las etiquetas
cv2_imshow(group_image)
```
